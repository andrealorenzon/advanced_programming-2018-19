We can use as basis $V_i = x^i$, or Vi = lagrange poly: (product (x - xj)/(xi-xj) when i!=j

The points are not decreasing. F is smooth (per assumption).

```python
def interpolation(myfun, q, x):
    """ Return polynomial interpolation of myfun at the points q, evaluated at the point x
        you can plot this by calling 
        
        plot(x, interpolation(myfun, q, x))"""
    N = len(q)
    I = arange(N)
    Qvec = vstack(q)  # whatever x 1 vector
    Xvec = vstack(x)
    
    B = Qvec**I   #Q is B matrix
    X = Xvec**I
    p = inv(B).dot(myfun(q))
    return x.dot(p)
```

how do we write the interpolation function to use Lagrange functions instead? 

```python
def lagrange_basis(q, x):
    """
    Takes a linspace of evaluation points x0...xN as q
    Takes a linspace of x points for plotting
    
    Return the Lagrange bases function evaluated at x"""
    N = len(q)                    # number of evaluation points
    I = arange(N)                 # vector of x0...xN points
    Qvec = q.reshape((-1,1))      # whatever x 1 vector
    Xvec = x.reshape((-1,1))      # whatever x 1 vector of all x linspace
    B = Qvec**I                   #Q is B matrix
    V = Xvec**I
    L = V.dot(inv(B))  # I would like to have a matrix here, to have V.dot(inv(B))
    return L
```
